```python
import torch
```

```python
N = 10
D_in = 1
D_out = 1
X = torch.randn(N,D_in)

liner_layer = torch.nn.Linear(in_features=D_in,out_features=D_out)

print(f"Layers Weight (W): {liner_layer.weight}\n")
print(f"Layer's Bias (b): {liner_layer.bias}\n")

y_hat_nn = liner_layer(X)

print(f"Output of nn.linear (first 3 rows ): \n{y_hat_nn[:3]}")

gelu = torch.nn.GELU()
sample = torch.randn(N)
activated_data = gelu(sample)
print(f"Sample Data: {sample}")
print(f"Activated Data: {activated_data}")
```
